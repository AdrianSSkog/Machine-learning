```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

m = 40
X = 6 * np.random.rand(m, 1) - 3
y = 0.5 * X**2 + X + 2 + np.random.randn(m, 1)

X_train, X_test = X[:30], X[30:] 
y_train, y_test = y[:30], y[30:]

plt.scatter(X_train, y_train)
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline

model = Pipeline([
   ("poly", PolynomialFeatures(degree=15)),
    ("lin_reg", LinearRegression())
])

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test).reshape(-1)

idx = np.argsort(X_test.reshape(-1))

plt.scatter(X_test, y_test)
plt.plot(X_test[idx], y_pred[idx], c="red")
plt.show()
```

```python
from sklearn.linear_model import Ridge, Lasso

model_ridge = Pipeline([
    ("poly", PolynomialFeatures(degree=15)),
    ("ridge", Ridge(alpha=100))
])

model_lasso = Pipeline([
    ("poly", PolynomialFeatures(degree=15)),
    ("lasso", Lasso(alpha=100))
])

model_ridge.fit(X_train, y_train)
model_lasso.fit(X_train, y_train)
```

```python
y_pr_rdge = model_ridge.predict(X_test)
y_pr_lass = model_lasso.predict(X_test)
y_pred = model.predict(X_test)

plt.scatter(X_train, y_train)
plt.scatter(X_test, y_test, c="orange")
plt.plot(X_test[idx], y_pred.reshape(-1)[idx], c="red")
plt.plot(X_test[idx], y_pr_rdge.reshape(-1)[idx], c="green")
plt.plot(X_test[idx], y_pr_lass.reshape(-1)[idx], c="purple")
plt.show()
```
